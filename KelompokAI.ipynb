Pengoptimalan Particle Swarm (PSO)

Beberapa studi tentang perilaku sosial kelompok hewan dikembangkan pada awal 1990-an. Penyelidikan ini mengungkapkan bahwa beberapa makhluk dalam kelompok tertentu, yaitu burung dan ikan, mampu mengirimkan pengetahuan di antara mereka sendiri dan bahwa kemampuan ini memberikan manfaat kelangsungan hidup yang signifikan bagi hewan-hewan ini. Terinspirasi oleh penelitian ini, Kennedy dan Eberhart mempresentasikan algoritma PSO pada tahun 1995, sebuah algoritma metaheuristik yang cocok untuk mengoptimalkan fungsi kontinu nonlinier. Algoritme ini terinspirasi oleh konsep kecerdasan kawanan, yang biasanya ditampilkan dalam pengelompokan hewan seperti kawanan dan kawanan.

Sebagaimana dinyatakan di awal, ikan atau sekawanan burung yang bergerak dalam kelompok ”dapat mengambil manfaat dari pengalaman semua anggota lainnya”. Dengan kata lain, jika seekor burung terbang secara acak mencari makanan, semua burung dalam kawanan dapat berbagi penemuan mereka dan membantu seluruh kawanan mendapatkan perburuan terbaik. Meskipun kita dapat meniru pergerakan sekawanan burung, kita juga dapat mengasumsikan bahwa setiap burung membantu kita dalam menemukan solusi optimal dalam ruang solusi berdimensi tinggi, dengan solusi terbaik yang ditemukan oleh kawanan menjadi solusi terbaik dalam ruang tersebut.


Menerapkan Pengoptimalan Particle Swarm menggunakan PySpwarms


PySwarms adalah alat berbasis Python untuk optimasi gerombolan partikel. Ini digunakan oleh para peneliti, praktisi, dan siswa kecerdasan segerombolan yang ingin menggunakan antarmuka tingkat tinggi deklaratif untuk menerapkan PSO pada masalah mereka. PySwarms menawarkan interaksi dengan optimasi swarm dan optimasi dasar dengan PSO.


PySwarms menerapkan teknik pengoptimalan gerombolan banyak partikel pada tingkat tinggi. Akibatnya, ia bercita-cita untuk menjadi ramah pengguna dan mudah beradaptasi. Modul pendukung juga dapat digunakan untuk membantu Anda mengatasi masalah pengoptimalan.


Di bagian ini, kami akan menerapkan pengoptimal terbaik global menggunakan API fungsional pyswarms.single.GBestPSO. Kami akan memplot fungsi dalam cara 2D dan 3D juga.


PySwarm dapat langsung diinstal oleh pip install pyswarms


Mengoptimalkan Fungsi Sphere

```python
# Import PySwarms
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
```

Kami akan bekerja untuk meningkatkan fungsi bola. Mari kita menempatkan beberapa pengaturan sewenang-wenang di pengoptimal kami untuk saat ini. Setidaknya, ada tiga langkah untuk optimasi:
1. Untuk mengonfigurasi swarm sebagai dictionary, atur hyperparameters.
2. Berikan dictionary dengan input yang relevan untuk membuat instance dari pengoptimal.
2. Panggil optimize() metode, dan beri tahu untuk menghemat biaya dan posisi terbaik dalam sebuah variabel.

```python
# Set-up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}
# Call instance of PSO
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)
# Perform optimization
cost, pos = optimizer.optimize(fx.sphere, iters=1000)
```

Ini akan menjalankan pengoptimal untuk 1000 iterasi sebelum mengembalikan biaya dan posisi terbaik kawanan.

Memvisualisasikan Fungsi


PySwarms menyertakan alat untuk memvisualisasikan perilaku kawanan Anda. Ini dibangun di atas matplotlib, menghasilkan bagan yang mudah digunakan dan sangat dapat disesuaikan. Modul plotters memiliki dua metode animasi: plot contour()dan plot surface(). Pendekatan ini, seperti namanya, memplot partikel dalam ruang 2-D atau 3-D.

Untuk memplot fungsi sphere, kita harus menambahkan mesh ke swarm kita. Hal ini memungkinkan kita untuk melihat di mana partikel berada dalam kaitannya dengan fungsi tujuan kita secara grafis. Menggunakan kelas Mesher, kita dapat mencapai ini.

```python
import matplotlib.pyplot as plt
from pyswarms.utils.plotters import plot_contour, plot_surface
from pyswarms.utils.plotters.formatters import Designer
from pyswarms.utils.plotters.formatters import Mesher
```

pyswarms.utils.plotters.formattersmodul berisi banyak formatters untuk menyesuaikan plot dan visualisasi. Selain Mesher, ada kelas Designer untuk memodifikasi ukuran font, ukuran gambar, dan sebagainya, serta kelas Animator untuk mengatur penundaan dan pengulangan animasi.

Plot 2-D

```python
m = Mesher(func=fx.sphere)
# Make animation
animation = plot_contour(pos_history=optimizer.pos_history,
                         mesher=m,
                         mark=(0,0)) # Mark minima
animation.save('mymovie.mp4')
```

Plot 3-D

```python
# preprocessing
pos_history_3d = m.compute_history_3d(optimizer.pos_history)
# adjust the figure
d = Designer(limits=[(-1,1), (-1,1), (-0.1,1)], label=['x-axis', 'y-axis', 'z-axis'])
# Make animation
animation3d = plot_surface(pos_history=pos_history_3d, 
                           mesher=m, designer=d,       
                           mark=(0,0,0))  # Mark minima
```
